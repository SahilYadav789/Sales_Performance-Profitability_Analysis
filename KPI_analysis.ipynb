```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from data import get_data

df = get_data()
df["Revenue"] = df["Price"] * df["Quantity"]
```

```python
total_revenue = df["Revenue"].sum()
print("Total Revenue:", total_revenue)
```

```python
total_orders = df["Order_ID"].nunique()
print("Total_Orders:", total_orders)
```

```python
aov = total_revenue / total_orders
print("Average Order Value:", aov)
```

```python
revenue_per_product = df.groupby("Product")["Revenue"].sum().reset_index()
pd.DataFrame(revenue_per_product)
```

```python
top_product_revenue = revenue_per_product["Revenue"].max()
top_product_contribution = (top_product_revenue / total_revenue) * 100
print(f"Dependency Risk : {top_product_contribution.round(2)}%")
```
