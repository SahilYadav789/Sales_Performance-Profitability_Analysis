* Project Title

Sales Performance & Profitability Analysis

* Project Description

Analyzed sales and product data to evaluate revenue, customer demand, and profitability. Performed data cleaning, feature engineering, KPI analysis, and time-based trend analysis. Built dashboards to present insights and provided data-driven recommendations for business growth.

* Tools & Skills

1. Python (Pandas, Matplotlib)
2. SQL concepts (Joins, Aggregations)
3. Data Analysis & EDA
3. KPI Analysis
4. Dashboard & Data Storytelling

* Key Achievements (Bullet Points)

1. Identified laptops as the top revenue and profit driver
2. Analyzed customer demand patterns showing mobiles as high-volume products
3. Performed month-on-month trend analysis to detect sales recovery
4. Calculated key KPIs including total revenue, profit, and AOV
5. Provided strategic recommendations to improve cross-selling and profitability

***Short Case Study on Analysis of Data***

**Business Problem**

The company wanted to understand its sales performance, customer demand, and profitability in order to make better business decisions and improve overall performance.

**Data Understanding**

The analysis used sales and product datasets containing order details, product categories, pricing information, and cost data for profit calculation.

**Analysis Performed**

* Cleaned and prepared the data, including date handling
* Performed feature engineering by creating revenue and profit metrics
* Analyzed product and category-level performance
* Conducted time-based trend analysis to identify growth and dips
* Calculated key business KPIs such as total revenue, profit, and AOV
* Built a dashboard to summarize insights for management

**Key Insights**

* Laptops are the top contributors to both revenue and profit
* Mobiles generate high demand and drive customer volume
* February experienced a temporary decline, followed by a strong recovery in March
* The business maintains a healthy balance between high-volume and high-value products

**Recommendations**

* Focus on laptops for profit maximization
* Use mobiles to attract and retain more customers
* Introduce desktops and CPUs to increase accessory sales
* Apply cross-selling strategies to improve overall revenue per customer
